```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['Gender'] = le.fit_transform(data['Gender'])

data['VisitFrequency'] = le.fit_transform(data['VisitFrequency'])

data['PreferredCuisine'] = le.fit_transform(data['PreferredCuisine'])

data['TimeOfVisit'] = le.fit_transform(data['TimeOfVisit'])

data['DiningOccasion'] = le.fit_transform(data['DiningOccasion'])

data['MealType'] = le.fit_transform(data['MealType'])
```

```python
x = data.drop(['HighSatisfaction'], axis = 1)
y = data['HighSatisfaction']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
print("x_train",x_train.shape)
print("x_test",x_test.shape)
print("y_train",y_train.shape)
print("y_test",y_test.shape)
```

```python
gbc = GradientBoostingClassifier(learning_rate=0.3, n_estimators=100, max_depth=3, min_samples_split=2, min_samples_leaf=1, ccp_alpha = 10 )

model = gbc.fit(x_train, y_train)

model
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

ass = accuracy_score(y_test, y_pred)*100

print(f"Accuracy_Score = {ass}")
```
